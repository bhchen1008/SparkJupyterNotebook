```python
from pyspark import SparkContext, SparkConf
conf = SparkConf().setAppName("WordCount").setMaster("local[*]")
sc = SparkContext(conf=conf)
```

```python
file = sc.textFile("/data/README.md")
```

```python
file.take(10)
```

```python
countsTmp = file.flatMap(lambda line: line.split(" "))
```

```python
countsTmp.collect()
```

```python
countsMap = file.map(lambda line: line.split(" "))
```

```python
countsMap.collect()
```

```python
countsTmp = countsTmp.map(lambda word: (word,1)) 
```

```python
countsTmp.take(10)
```

```python
counts = countsTmp.reduceByKey(lambda a,b: a+b)
```

```python
counts.collect()
```
